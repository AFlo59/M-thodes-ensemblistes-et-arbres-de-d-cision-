```python
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier, Pool
from sklearn.metrics import accuracy_score

''' 

pclass: Classe du billet (1re, 2e, 3e).
survived: Variable binaire indiquant si le passager a survécu (1) ou non (0).
name: Nom du passager.
sex: Sexe du passager.
age: Âge du passager.
sibsp: Nombre de frères et sœurs ou conjoints à bord.
parch: Nombre de parents ou d'enfants à bord.
ticket: Numéro du billet.
fare: Prix du billet.
cabin: Numéro de cabine.
embarked: Port d'embarquement (C = Cherbourg, Q = Queenstown, S = Southampton).
boat: Numéro du canot de sauvetage.
body: Numéro d'identification du corps (en cas de décès).
home.dest: Destination ou lieu d'origine du passager. '''







df = pd.read_csv('/home/utilisateur/Documents/dev ia/M-thodes-ensemblistes-et-arbres-de-d-cision-/datasets/titanic.csv')



# Liste des valeurs nulles par colonne
valeurs_nulles_par_colonne = df.isnull().sum()
# print(valeurs_nulles_par_colonne)

# Choix des colonnes 
df = df[['survived', 'pclass', 'sex', 'age',]]


# Supprimer les lignes contenant des valeurs nulles
df_sans_nulles = df.dropna()
# print(df_sans_nulles.shape)

# conversion des floats cat en str
df_sans_nulles.loc[:, 'pclass'] = df_sans_nulles['pclass'].astype(str)


X = df_sans_nulles.drop('survived', axis=1)
y = df_sans_nulles[['survived']]




# division du dataset en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.2, random_state=42)


# print(X.dtypes)

# Définir les colonnes catégorielles (indices des colonnes catégorielles)
cat_features_indices = [0,1]

# Créer un objet Pool pour les données d'entraînement en spécifiant les catégories
train_pool = Pool(data=X_train, label=y_train, cat_features=cat_features_indices)

# Créer un modèle CatBoostClassifier
model = CatBoostClassifier(iterations=100, depth=5, learning_rate=0.1, eval_metric='Accuracy')


# Entraîner le modèle sur l'objet Pool
model.fit(train_pool, verbose=False)

# Faire des prédictions sur l'ensemble de test
y_pred = model.predict(X_test)



# Accéder à l'accuracy sur l'ensemble d'entraînement
training_accuracy = model.get_best_score()['learn']['Accuracy']

# Afficher les accuracies
print(f"Training Accuracy: {training_accuracy:.4f}")

# Calculer l'accuracy sur l'ensemble de test
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
# regression logistique 

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder

# Charger les données
df = pd.read_csv('/home/utilisateur/Documents/dev ia/M-thodes-ensemblistes-et-arbres-de-d-cision-/datasets/titanic.csv')

# Choix des colonnes
df = df[['survived', 'pclass', 'sex', 'age']]

# Supprimer les lignes contenant des valeurs nulles
df_sans_nulles = df.dropna()

# Conversion des floats cat en str
df_sans_nulles['pclass'] = df_sans_nulles['pclass'].astype(str)

# Encoder la colonne 'sex' si nécessaire (décommentez si vous souhaitez l'encoder)
label_encoder = LabelEncoder()
df_sans_nulles['sex'] = label_encoder.fit_transform(df_sans_nulles['sex'])

# Séparer les features (X) de la target variable (y)
X = df_sans_nulles.drop('survived', axis=1)
y = df_sans_nulles['survived']

# Division du dataset en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.2, random_state=42)

# Créer un modèle de régression logistique
model = LogisticRegression()

# Entraîner le modèle sur l'ensemble d'entraînement
model.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred = model.predict(X_test)

# Mesurer l'accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Afficher le rapport de classification
print("Classification Report:\n", classification_report(y_test, y_pred))
```
